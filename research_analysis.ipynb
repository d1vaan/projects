## Предобработка данных

### Обработка пропусков

Для начала подсчитаем общее количество пропусков по столбцам:

```python
data.isna().sum() # подсчет суммы пропусков
```

Учитывая, какие данные мы будем анализировать в дальнейшем, можем пропустить заполнение пропусков в столбцах: `airport_nearest`, `park_around3000`, `park_nearest`, `pounds_around3000`, `park_nearest`. 

Из-за необходимости в дальнейшем исследовании столбцов `locality_name` , заполним пропуски в столбце, заменив их на `Неизвестно`. В случае со столбцами `cityCenters_nearest`и `days_exposition` оставим пропуски, поскольку оба столбца содержат более `10%` пропусков, заполнение которых в будущем может повлиять на результаты исследования.

Пропуски в столбцах `balcony` и `is_apartment` заполним значениями `0`.

Все остальные значения заполним медианными значениями из столбца с учетом особенностей данных параметров.

```python
data['locality_name'] = data['locality_name'].fillna( # заполнение пропусков с столбце `locality_name`
    'Неизвестно'                                      # строкой 'Неизвестно'
)
```

```python
data['balcony'] = data['balcony'].fillna(0) # заполнения пропусков в столбце `balcony` нулями
data['is_apartment'] = data['is_apartment'].fillna(0) # заполнения пропусков в столбце `is_apartment` нулями
```

```python
data['ceiling_height'] = data['ceiling_height'].fillna( # заполнение столбца `ceiling_height` медианными значениями в нём
    data['ceiling_height'].median()
)
```

Заполнив простые пропуски, заполним теперь более сложные для заполнения пропуски: `living_area`, `kitchen_area`. 

Высчитаем медианное соотношение общей площади ( в столбце этих праметров отсутствуют пропуски ) к площади этих параметров. Заполним пропущенные значения в соответствии с медианным значением соотношениях этих двух параметров.

Создадим для этого функцию:

```python
# задаём функцию `fill_skip` с параметром column:
def fill_skip(column):
# высчитываем медиану отношения общей площади к прощади выбранного праметра:
    median_value = (data[column] / data['total_area']).median()
# заполняем пропуски, перемножая общую площадь квартиры на медианное значение
    data[column] = data[column].fillna(
        data['total_area'] * median_value
    )
```

```python
fill_skip('kitchen_area') # заполняем пропуски столбца `kitchen_area`

fill_skip('living_area') # заполняем пропуски столбца `ceiling_area`
```

Удостоверимся в заполнение пропусков:

```python
data.isna().sum() # подсчет суммы пропусков
```

### Изменение типа данных

Ещё раз выведем информацию о таблице, что бы вспомнить типы данных каждого столбца:

```python
data.info() # вывод общей информации о таблице
```

В первую очередь изменим тип и переведём в более читаемый вид столбец `first_day_exposition`. Для удобства чтения данных, оставив только дату размещения объявления:

```python
# изменяем тип данных столбца `first_day_exposition` на datetime
data['first_day_exposition'] = pd.to_datetime(
    data['first_day_exposition'], format='%Y-%m-%d %H:%M'
)
# округляем даты размещения объявления до дней
data['first_day_exposition'] = data['first_day_exposition'].dt.floor('1D') 
```

Изменим так же некоторым столбцам тип данных. В случае со столбцом `is_apartment`, он требует булевый тип данных, поскольку квартира может являться или не являться апартаментами.

```python
# меняем тип данных столбца `is_apartment` на булевый
data['is_apartment'] = data['is_apartment'].astype('bool')
```

Убедимся в изменение типа данных, снова выведем информацию о таблице:

```python
data.info()
```

## Расчёты и добавление результатов в таблицу

Для дальнейших исследовательских рассчетов нам потребуются такие параметры, как: `цена за квадратный метр`; `день недели`, `месяц` и `год` публикации; `этаж квартиры` в формате `'Первый', 'Последний', 'Другой'`; соотношения площадей `жилой к общей` и  `кухни к общей`.

Посчитаем все эти параметры:

```python
# высчитаем и запишем в столбец 'price_per_meter' цену за квадратный метр
data['price_per_meter'] = data['last_price'] / data['total_area']
# округлим значения столбца для лучшей читаемости
data['price_per_meter'] = data['price_per_meter'].astype('int')

# высчитываем отношение жилой площади квартиры к общей, записываем все в столбец 'living_to_total'
data['living_to_total'] = data['living_area'] / data['total_area']

# высчитываем отношение площади кухни к общей площади, записываем все в столбец 'kitchen_to_total'
data['kitchen_to_total'] = data['kitchen_area'] / data['total_area']
```

Для параметра `день недели` и категоризации этажа квартиры требуется написать функцию для дальнейшего применения её на методе `apply`. Дни недели запишем в строчном виде для лучшей читаемости данных. 

Сделаем подготовительные этапы: расчитаем промежуточные значения дней недели в числовом формате и создадим обе функции.

```python
# создаём промежуточные числовые значения дней недели
data['weekday'] = data['first_day_exposition'].dt.weekday.astype('int')
```

```python
# создаём функцию <weekday> для записи дня недели в строчном виде
# на вход функция получает числовое значение дня недели:
def weekday(day):
# если числовое значение дня недели 0, то:
    if day == 0:
# день недели - 'monday'
        return 'monday'
# если числовое значение дня недели 1, то:
    if day == 1:
# день недели - 'tuesday'
        return 'tuesday'
# если числовое значение дня недели 2, то:
    if day == 2:
# день недели - 'wednesday'
        return 'wednesday'
# если числовое значение дня недели 3, то:
    if day == 3:
# день недели - 'thursday'
        return 'thursday'
# если числовое значение дня недели 4, то:
    if day == 4:
# день недели - 'friday'
        return 'friday'
# если числовое значение дня недели 5, то:
    if day == 5:
# день недели - 'saturday'
        return 'saturday'
# в ином случае день недели - 'sunday'
    return 'sunday'
```

```python
# создаём функцию <floor> для категоризации этажа
# на вход поступает номер этажа квартиры
def floor(row):
# если этаж квартиры первый, то:
    if row['floor'] == 1:
# присуждается категория 'первый'
        return 'первый'
# если этаж квартиры равен общей этажности дома, то:
    if row['floor'] == row['floors_total']:
# присуждается категория 'последний'
        return 'последний'
# в ином случае присуждается категория 'другой'
    return 'другой'
```

Проведя подготовительные этапы, можно теперь спокойно расчитывать оставшиеся параметры:

```python
# применим метод 'apply' с функцией 'floor' для нового столбца 'floor_type':
data['floor_type'] = data.apply(floor, axis=1)
```

```python
# применим метод 'apply' с функцией 'weekday' для столбца 'weekday':
data['weekday'] = data['weekday'].apply(weekday)

# выделем номер месяца, в котором было размещено объявление в отдельный столбец
data['month'] = data['first_day_exposition'].dt.month

# расчитаем год, в котором было размещено объявление в отдельный столбец
data['year'] = data['first_day_exposition'].dt.year
```

```python
# строим гистограмму по высоте потолков
# 
data.hist(
    'ceiling_height', 
    bins=100,
    figsize=(12, 6)
)
```

### Центральный район города

Нам предстоит вычислить центральную зону города. Для начала сделаем срез данных на основе датасета. Далее видоизменим столбец `cityCenters_nearest`, расчитав целочисленное удаление от центра в километрах. А после построим график зависимости цены от удалённости от центра.

```python
# создаём таблицу spb_data по срезу из старого датасета
spb_data = data.query('locality_name == "Санкт-Петербург"')
```

Поскольку обработка столбца `cityCenters_nearest` невозможна из-за пропусков, подсчитаем кол-во пропусков в столбце от общего кол-ва данных и решим, что с ними делать.

```python
# делим кол-во пропусков в столбце 'cityCenters_nearest' на общее кол-во строк
spb_data['cityCenters_nearest'].isna().sum() / spb_data['last_price'].count()
```

Как можем заметить, пропусков лишь `3%`. По этой причине удаление или заполнение их средними значениями слабо повлияют на расчеты. 

Для облегчения дальнейшей работы с таблицей, заполним пропуски медианным значением из столбца.

```python
# заполнение столбца `cityCenters_nearest` медианными значениями в нём
pd.set_option('chained_assignment', None)
spb_data['cityCenters_nearest'] = spb_data['cityCenters_nearest'].fillna( 
    spb_data['cityCenters_nearest'].median()
)
```

Теперь создадим столбец `center_nearest`, где будет храница информация об удалённости от центра в км.


```python
pd.set_option('chained_assignment', None)
spb_data['center_nearest'] = ( spb_data['cityCenters_nearest'] / 1000 ).astype('int')
```

Теперь, обработав данные для удобства работы, найдём крайнюю точку центрального района.

```python
(
    spb_data
    .pivot_table(index='center_nearest', values='last_price', aggfunc='mean')
    .plot(
    style='o-',
    grid=True, 
    figsize=(12, 6)
    )
)
```

#### Вывод

Как мы можем увидить по графику, резкий и пологий график зависимости средней цены от удалённости от центра наблюдаются вплоть до 9км. Из этого можно сделать вывод, что центром города можно считать область в радиусе `9км`.
